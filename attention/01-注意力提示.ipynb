# Query、Key、Value



我们可以把注意力机制想象成一台自动售货机。假设这台自动售货机有很多商品，每次只能购买一件商品，但有些商品对你来说更重要一些。那么你可以通过注意力机制来调节自己关注的重点。

比如说，如果你想要买一罐可乐，你可以把自己的注意力放在可乐的位置，这样就可以更快地找到它，并成功买到可乐。而如果这时候突然出现了一件更重要的事情，比如有一只猫跑进来了，你可以把注意力转向猫的位置，从而做出更有效的反应。

同样的道理也适用于人工智能领域。当一个AI系统接收到大量的数据时，它可以使用注意力机制来调整自己的关注点，从而更好地处理这些数据。这样就可以让AI系统变得更加聪明、更加智能了！

在注意力机制中，query、key和value都是用来描述一组数据的概念，它们之间的关系非常密切。

首先，query表示的是某个特定任务的需求或意图，它可以被看作是一种“提问”的方式；而key则表示每一条数据的重要性程度，它可以被看作是一种“评分”或者“标记”的方式；最后，value则表示每一条数据的实际内容，它可以被看作是一种“答案”或者“结果”的方式。

因此，在实际应用过程中，query可以用来筛选出符合要求的数据，key可以用来对筛选出来的数据进行排序，而value则可以直接用来获取最终的结果。这样就可以有效地提高系统的效率，实现更好的性能表现了！

```python
# 注意力可视化

# show_heatmaps：将输入的矩阵转换为热图，其输入matrices的形状是 （要显示的行数，要显示的列数，查询的数目，键的数目）。

attention_weights = torch.eye(10).reshape((1, 1, 10, 10))
d2l.show_heatmaps(attention_weights, xlabel='Keys', ylabel='Queries')
```

```python
# 随机生成一个10x10的矩阵并使用softmax运算来确保每行都是有效的概率分布，然后可视化输出注意力权重。

from torch.nn import functional as F
X = torch.rand((10,10))
X = F.softmax(X,dim=1)
X = X.reshape((1,1,10,10))
d2l.show_heatmaps(X,xlabel='Columns',ylabel='Rows')
```
