```python
import json
from openai import OpenAI
from typing import Dict, List, Optional, Tuple, Union
import itertools
from tqdm import tqdm
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
```

```python
def chat(prompt: str, history: List=[]) -> str:
    client = OpenAI()   
    history.append({'role': 'user', 'content': prompt})
    response = client.chat.completions.create(
        model='gpt-4-0125-preview',
        messages=history,
        temperature=1
    )
    return response.choices[0].message.content
```

```python
with open('./huanhuan.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
def get_input(data: List):
    return [f"Q:{item['instruction']}\nA:{item['output']}" for item in data]
```

```python
data = get_input(data)
```

```python
# 使用itertools.combinations获取所有的组合
combinations = itertools.combinations(data[:5], 2)
```

```python
# 创建一个新的列表来保存结果
new_list = []
for combination in combinations:
    # 使用join函数将组合中的元素连接成一个字符串，并添加到新的列表中
    new_list.append("\n\n".join(combination))
```
